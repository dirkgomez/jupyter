## Daten Oktoberfest 

```python
import pandas as pd
import seaborn as sns

data_url = 'https://www.opengov-muenchen.de/dataset/8d6c8251-7956-4f92-8c96-f79106aab828/resource/e0f664cf-6dd9-4743-bd2b-81a8b18bd1d2/download/oktoberfestgesamt19852019.csv'

df=pd.read_csv(data_url)

beer_mean = df["bier_konsum"].mean()
chicken_mean = df["hendl_konsum"].mean()

df["beer_percent"] = (df['bier_konsum'] / beer_mean) * 100
df["chicken_percent"] = (df['hendl_konsum'] / chicken_mean) * 100

sns.lineplot(data=df, x="jahr", y="beer_percent", label="Durst")
p = sns.lineplot(data=df, x="jahr", y="chicken_percent", label="Hunger")

p.set_xlabel("Jahr")
p.set_ylabel("Konsum in % zum Mittel")
```
